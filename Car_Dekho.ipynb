```python
#Q1)From which manufacturing year to which manufacturing year vehicles are present in this data ? 
max_year = df['Year'].max() #stores the maximum value of year.
min_year = df['Year'].min() #stores the minimum value of year.
```

```python
print("Cars are present from year " + str(min_year) + " to " + str(max_year))
```

```python
#Q2)What is the lowest price to which a vehicle is sold ?
min_price = df['Selling_Price'].min()
print("The lowest price is " + str(min_price) + " lakhs")
```

```python
#Q3)What is the highest price to which a vehicle is sold ?
max_price = df['Selling_Price'].max()
print("The highest price is " + str(max_price) + " lakhs.")
```

```python
#Q4)How many records are there in this data ? 
records = df.shape[0]
print("The total number of records are " + str(records))
```

```python
#Q5)Are there any missing records in this data ? 
#Basically missing records mean Null Values (NaN or None)

missing_val = df.isnull().sum()
print("The number of missing values are: \n\n" + str(missing_val))
```

```python
#Q6)How many different vehicles are present in this data ? 
diff_veh = df["Car_Name"].nunique()
print("The number of different vehicles are " + str(diff_veh))
```

```python
#Q7)Which is the most sold vehicle in this data ? 
most_sold_car = df['Car_Name'].value_counts().idxmax()
num_of_times = df['Car_Name'].value_counts().max()
print("The most sold car is " + str(most_sold_car) + ". And it is sold " + str(num_of_times) + " times.")
```

```python
#Q8)Does the database include any CNG vehicle ? If yes how many of them are there ?
CNG = (df["Fuel_Type"] == "CNG").sum()
print("The number of CNG vehicles are " + str(CNG))
```

```python
#Q9)How many vehicles here are for sell from Individuals directly ? 
seller_type= df["Seller_Type"].value_counts()

indi_count = seller_type["Individual"]
print("Number of individual sellers are " + str(indi_count))
```

```python
#Q10)Does this database contain auto transmission vehicles ? If yes how many of them are there ?
auto = (df["Transmission"] == "Automatic").sum()
print("The number of automatic vehicles are " + str(auto))
```

```python
#Q11)How many single person owned vehicles are there in this database ? 
single = (df["Owner"] == 1).sum()
print("The number of single person owned vehicles are " + str(single))
```

```python
#Q12)Which is the most and least cost depreciated vehicle in data ? 
df["Depreciation"] = ((-(df["Selling_Price"]) + df["Present_Price"])/df["Present_Price"])*100  #this creates a new column for depreciation in the dataframe
```

```python
#taking depreciation as Present_Price - Selling_Price is not advisable as for cars/two_wheelers having less price, the depreciation would always come out
#to be less. So it is better to take the percentage decrease from the present_price to get a better idea.
```

```python
#Q17)Can we find out data of only two wheelers from this data ?
#we can do an assumption that all two wheelers are less than 3.5 lakh rupees
two_wheelers = df[df['Present_Price'] < 3.5]
two_wheelers.head()
```

```python
two_wheelers.reset_index(drop=True,inplace=True) #changes the index
two_wheelers.head() 
```

```python
#alto k10, 800 and omni are not two wheelers, so we will remove these from the new dataframe
```

```python
two_wheelers = two_wheelers.loc[two_wheelers.Car_Name != "alto k10"]
two_wheelers = two_wheelers.loc[two_wheelers.Car_Name != "800"]
two_wheelers = two_wheelers.loc[two_wheelers.Car_Name != "omni"]
two_wheelers.head()
```

```python
two_wheelers.reset_index(drop=True,inplace=True) #changes the index
two_wheelers
```
